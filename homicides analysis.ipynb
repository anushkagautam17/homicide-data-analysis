```python
df.replace("Unknown", np.nan, inplace=True)
```

```python
df['Victim Sex'].fillna("Unknown", inplace=True)
df['Perpetrator Sex'].fillna("Unknown", inplace=True)
df['Weapon'].fillna("Unknown", inplace=True)
df['Relationship'].fillna("Unknown", inplace=True)
```

```python
def clean_weapon(x):
    if pd.isnull(x):
        return "Unknown"
    x = x.lower()
    if any(k in x for k in ["gun", "firearm", "rifle", "shotgun"]):
        return "Firearm"
    elif "knife" in x:
        return "Sharp Object"
    elif "blunt" in x:
        return "Blunt Object"
    else:
        return "Other"

df['Weapon_Group'] = df['Weapon'].apply(clean_weapon)
```

```python
def clean_relationship(x):
    if pd.isnull(x):
        return "Unknown"
    
    x = x.lower()
    
    if any(k in x for k in ["wife","husband","girlfriend","boyfriend"]):
        return "Partner"
    elif any(k in x for k in ["father","mother","son","daughter"]):
        return "Family"
    elif any(k in x for k in ["brother","sister"]):
        return "Sibling"
    elif any(k in x for k in ["friend","neighbor","acquaintance"]):
        return "Known"
    elif "stranger" in x:
        return "Stranger"
    else:
        return "Other"

df['Relationship_Group'] = df['Relationship'].apply(clean_relationship)
```

```python
#to convert into numeric as there are string in age columns

df['Victim Age'] = pd.to_numeric(df['Victim Age'], errors='coerce') 
df['Perpetrator Age'] = pd.to_numeric(df['Perpetrator Age'], errors='coerce')
```
